```python
import pandas as pd
import numpy as np
```

```python
file = "./static/data/hcny_2019.csv"
b_2019 = pd.read_csv(file)
b_2019.head()
```

```python
b2019= b_2019.rename(columns={"Q_Group": "Quarter"})
b2019.head()
```

```python
b2019.dropna(subset=['Year', 'Quarter','State','Colonies','Max Colonies','Lost Colonies', 'Added Colonies', 'Renovated Colonies'], inplace=True)
b2019.head()
```

```python
### Change numbers to quarter identification
b2019['Quarter'].replace(to_replace = 5, value= "Q1", inplace=True)
b2019['Quarter'].replace(to_replace = 1, value= "Q2", inplace=True)
b2019['Quarter'].replace(to_replace = 7, value= "Q3", inplace=True)
b2019['Quarter'].replace(to_replace = 8, value= "Q4", inplace=True)
b2019.head()
```

```python
file = "./static/data/hcny_2018.csv"
b_2018 = pd.read_csv(file)
b_2018.head()
```

```python
b_2018.dropna(subset=['Year', 'Quarter','State','Colonies','Max Colonies','Lost Colonies', 'Added Colonies', 'Renovated Colonies'], inplace=True)
# b_2018.head()

### Change numbers to quarter identification
b_2018['Quarter'].replace(to_replace = 5, value= "Q1", inplace=True)
b_2018['Quarter'].replace(to_replace = 1, value= "Q2", inplace=True)
b_2018['Quarter'].replace(to_replace = 7, value= "Q3", inplace=True)
b_2018['Quarter'].replace(to_replace = 8, value= "Q4", inplace=True)
b_2018.head()
```

```python
df = b2019.append(b_2018, ignore_index = True)
df.tail()
```

```python
file = "./static/data/hcny_2016.csv"
b_2016 = pd.read_csv(file)
b_2016.head()
```

```python
b_2016.dropna(subset=['Year', 'Quarter','State','Colonies','Max Colonies','Lost Colonies', 'Added Colonies', 'Renovated Colonies'], inplace=True)
# b_2018.head()

# ### Change numbers to quarter identification
b_2016['Quarter'].replace(to_replace = 5, value= "Q1", inplace=True)
b_2016['Quarter'].replace(to_replace = 1, value= "Q2", inplace=True)
b_2016['Quarter'].replace(to_replace = 7, value= "Q3", inplace=True)
b_2016['Quarter'].replace(to_replace = 8, value= "Q4", inplace=True)
b_2016.head()
```
